```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python

# Define filter coefficients (b: numerator, a: denominator)
bint = [1.0]
aint = [1.0,-1.0]
b = [1.0, -1.3, 0.3825]#[0.3825, -1.3, 1.0]
a = [1.0, -1.995, 0.995]#[0.995,   -1.995, 1.0]

# Check filter stability
z, p, k = signal.tf2zpk(b,a)
print("Poles:", p)
print("Zeros:", z)
print("Gain:", k)

z, p, k = signal.tf2zpk(bint,aint)
print("Poles:", p)
print("Zeros:", z)
print("Gain:", k)


# Check frequency response
w, h = signal.freqz(b, a)
wint, hint = signal.freqz(bint, aint)

# g = np.sum(np.abs(hint/(1.+hint)))/np.sum(np.abs(h/(1.+h)))
# print(g, np.sum(np.abs(hint/(1.+hint))))

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.plot(w, np.abs(1/(1.+h)))
plt.plot(wint, np.abs(1/(1.+hint)))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.yscale('log')
plt.xscale('log')
plt.grid()

plt.subplot(1,2,2)
plt.plot(w, np.abs(h/(1.+h)))
plt.plot(wint, np.abs(hint/(1.+hint)))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.yscale('log')
plt.xscale('log')
plt.grid()
```
